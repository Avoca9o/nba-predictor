Пункт 9 - в новом чистом блокноте выполнить загрузку обученной модели из эксперимента в mlflow с тегом “PRD” и сделать тестовый предикт


```python
import mlflow
import pandas as pd
```

```python
mlflow.set_tracking_uri("http://127.0.0.1:5000")
client = mlflow.tracking.MlflowClient()

runs = client.search_runs(
    experiment_ids=['1'],
    filter_string="tags.prd = 'true'",
    order_by=["start_time desc"]
)

if runs:
    run_id = runs[0].info.run_id
    
    model_uri = f"runs:/{run_id}/model"
    model = mlflow.sklearn.load_model(model_uri)
    
    expected_features = model.feature_names_in_
    df = pd.read_csv('model/features_engineered.csv')
    X_test = df[expected_features].dropna().iloc[-1000:]
    
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]
    
    print(f"Распределение: 0={sum(y_pred==0)}, 1={sum(y_pred==1)}")
    
else:
    print("Модель с тегом PRD не найдена")
```
