### Просто вспомогательная табличка, для того чтобы можно было маппить аббревиатуру с названием команды на ее id в датасете

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/line_score.csv')
df.head()
```

```python
abbreviations = df['team_abbreviation_home'].unique()

rows = []
for abbreviation in abbreviations:
    team_id = df.loc[df['team_abbreviation_home'] == abbreviation, 'team_id_home'].iloc[-1]
    rows.append({'abbreviation': abbreviation, 'id': team_id})

output_df = pd.DataFrame(rows)
output_df.to_csv('data/abbr_id_mapper.csv', index=False)

```
