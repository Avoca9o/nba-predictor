```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('result.csv')
display(df.head())
display(df.info())
display(df.describe())
```

```python
corr_matrix = df.corr()
spearman_corr_matrix = df.corr(method='spearman')
kendall_corr_matrix = df.corr(method='kendall')
display(corr_matrix)
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(spearman_corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Spearman Correlation Matrix')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(kendall_corr_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Kendall Correlation Matrix')
plt.show()
```

```python
upper = corr_matrix.where(
    np.triu(np.ones(corr_matrix.shape), k=1).astype(bool)
)

result = [
    (col, row, upper.loc[row, col])
    for col in upper.columns
    for row in upper.index
    if pd.notnull(upper.loc[row, col]) and abs(upper.loc[row, col]) > 0.9
]

for feat1, feat2, value in result:
    print(f'{feat1} - {feat2}:  = {value:.2f}')
```
